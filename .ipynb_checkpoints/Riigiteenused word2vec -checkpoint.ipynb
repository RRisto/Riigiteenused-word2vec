## Mudeli ehitamine

Kui tahad teada, kus kohast andmed võtsin ja mida nendega tegin, siis loe selle kohta [siit](https://github.com/RRisto/Riigiteenused-word2vec). Mudeliks kasutan word2vec algoritmi. Selle salajasemad hingesopid on mulle veel tundmatud, mistõttu ei klassifitseeru siinsed järeldused eksperdi arvamuseks. Kuna tekst, mida analüüsin, on väike, ei saagi mudeli järeldused olla väga kindlad. Hakkame mudelit ehitama. Esimese sammuna teeme fraasid (sõnad, mis tihti esinvead koos nt Los Angeles asemele teeb Los_Angeles).

```python
import word2vec
word2vec.word2phrase('riigiteenused_mudelisse.txt', 'mudel/riigiteenused-phrases', verbose=True)   
```

Eelmise sammu võib ka vahele jättan ning otse toorteksti mudelisse sööta. Treenime mudeli.

```python
word2vec.word2vec('mudel/riigiteenused-phrases', 'mudel/riigiteenused.bin', size=100, verbose=True)
```

See genereeris "riigiteenused.bin" faili, mis sisaldab sõnavectoreid binaarformaadis. Nüüd teeme vektorite klastedamise treenitud mudeli põhjal.

```python
word2vec.word2clusters('riigiteenused_mudelisse.txt', 'mudel/riigiteenused-clusters.txt', 100, verbose=True)
```

See lõi faili "riigiteenused-clusters.txt", mis iga klastreid iga sõna jaoks. Laeme mudeli üles ning hakkame sellega töötama.

```python
model = word2vec.load('mudel/riigiteenused.bin')
```

Nüüd saame näiteks iga sõna koosinuse sarnasuse leida.

```python
indexes, metrics = model.cosine('teenus')
model.generate_response(indexes, metrics).tolist()
```

## Analoogiad

Nüüd läheb põnevamaks. Saame teha lihtsat aritmeetikat sõnadega. Mis jääb sõnast "õigus" alles kui sellest maha lahutada "elanik" ning liita sõna "riik".

```python
indexes, metrics = model.analogy(pos=['õigus', 'riik'], neg=['elanik'], n=10)
indexes, metrics
model.generate_response(indexes, metrics).tolist()
```

```python
indexes, metrics = model.analogy(pos=['õigus', 'riik'], neg=['elanik'], n=10)
indexes, metrics
model.generate_response(indexes, metrics).tolist()
```

Kõige sobivam vastus tundub olevat välismaalane, mis tundub isegi intuitiivselt sobivat. Riiklik õigus, mis pole seotud elanikuga võiks ju olla seotud välismaalasega. Sellise aritmeetika tegemine on siiski suhteliselt vähe intuitiivne. Nüüd vaatame, mis sõnad on samas klastris terminiga "teenus".

```python
clusters = word2vec.load_clusters('mudel/riigiteenused-clusters.txt')
clusters[b'teenus']
clusters.get_words_on_cluster(90).shape
clusters.get_words_on_cluster(90)[:10]
```

```python
Tulemused on suhtleiselt segased, kuid tuleb arvestada, et teksti hulk on suhteliselt väike.
```

## Visualiseerimine 

Visualiseerimine sõnade sarnasusi ja erisusi. Selleks loome funktsiooni, mis võtab igast sõna kohta mudelist numbrite vektori, teeb neile põhikomponendi analüüsi (PCA), et saada 2-dimensioonilised andmed, mida on võimalik visualiseerida.

```python
from matplotlib import pylab
from sklearn.decomposition import PCA

def plot(words):
    %matplotlib inline
    embeddings = [model[w] for w in words]
    
    pca = PCA(n_components=2) 
    two_d_embeddings = pca.fit_transform(embeddings)
    
    pylab.figure(figsize=(5,5))  # in inches
    for i, label in enumerate(words):
        x, y = two_d_embeddings[i,:]
        pylab.scatter(x, y)
        pylab.annotate(label, xy=(x, y), xytext=(5, 2), textcoords='offset points', ha='right', va='bottom')
        pylab.xlabel("Komponent 1")
        pylab.ylabel("Komponent 2")
    pylab.show()
```

Nüüd vaatame visuaalselt, kui sarnased mõned huvipakkuvad sõnad on.

```python
plot("teenus taotlus luba tegevusluba litsents päring registreerimine esitamine sertifikaat tõend kontroll esita kontrollimine väljastamine".split())
```

Graafikult on näha, et "teenus" on väga lähedal sõnale "päring". "Luba" ja "tõend" on samuti üksteisele väga lähedal. Samuti on "sertifikaat" lähedal sõnale "esitamine", mis on loogiline, kuna sertifikaate tuleb tihti millegi tõendamiseks esitada. Proovime veel sõnu.

```python
plot("teenus taotlus luba tegevusluba litsents päring registreerimine esitamine sertifikaat tõend kontroll saa esita väljastamine õigus kohustus andmed eesti riik isik abielu avalik dokument taotlemine".split())
```

Jooniselt on näha, et "riik" on väga lähedal sõnadele "kotroll" ja "päring" ning ka sõnale "andmed". Sõnad "väljastamine", "esitamine", "taotlemine", "esita" on üksteisele lähedal, mis on loogiline, kuna need tegevused käivad üldjuhul koos. Proovime veel sõnu.

```python
plot("teenus õigus kohustus riik kodanik ettevõtja".split())
```

Jooniselt on näha, et "riik" on lähemal sõnale "kohustus", kui sõnale "teenus" (x-telje suhtes vaadatud on ta midagi vahepealset). Sõnad "kodanik" ja "ettevõtja" on lähemal mõistele "õigus" kui muudele mõistetele. 

```python
plot("riik kohustus õigus kontroll kontrollimine isik teenus".split())
```

Siin on jooniselt näha, et "teenus" on pigem lähemal mõistetele "kontroll" ja "kontrollimine" kui mõistetele "õigus" ja "riik". "Riik" on pigem lähemal mõistele "kohustus", kui mõistele "teenus". Vaadates asju y-telje poolt, siis on "riik" mõistete "isik" ja "õigus" vahel. Ning pärast mõisteid "õigus" järgnevad "kontroll" ja "teenus". Võime seda harjutust jätkata, kuid loodetavasti saite poindile pihta. Tegemist on huvitava võimalusega sõnadevahelisi seoseid visualiseerida ning sellest lugeda välja huvitavaid seoseid.
